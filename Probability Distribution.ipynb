## Generate a normal distribution of size 2*3 with mean at 1 and standard deviation of 2

```python
from numpy import random
x = random.normal(loc = 1, scale = 2, size = (2,3))
print(x)
```

## Visualization of normal distribution

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(random.normal(size=1000), hist=False)
plt.show
```

```python
data = random.normal(size = 100)
sns.kdeplot(data,bw_adjust = 0.5)
plt.show
```

## BINOMIAL DIASTRIBUTION

## Given 10  trials for coin toss, generate the 10 data points

```python
x = random.binomial(n = 10, p = 0.5, size = 10)
print(x)
```

```python
# n represents the number of trials, p is the probability of occurance
# size is the shape of the returned array
```

## VISUALIZATION OF NORMAL DISTRIBUTION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
binomial_data = random.binomial(n=10, p=0.5, size = 1000)
sns.histplot(binomial_data, bins = 10, kde = False, color = 'yellow')
plt. title ("A binomial Distribution")
plt.xlabel ('Number of success')
plt.ylabel ('Frequency')
plt.show
```

```python
binomial_data = random.binomial(n=10, p=0.5, size = 1000)
sns.histplot(binomial_data, bins = 10, kde = True, color = 'Yellow')
plt. title ("A binomial Distribution")
plt.xlabel ('Number of success')
plt.ylabel ('Frequency')
plt.show
# we changed kde to true
```

## Difference btn a normal and a binomial distribution

```python
normal_distribution = random.normal(loc = 50, scale = 5, size = 1000)
binomial_distribution = random.binomial(n = 100, p = 0.5, size = 1000)
sns.kdeplot(normal_distribution, label="Normal", color='purple')
sns.kdeplot(binomial_distribution, label="Binomial", color='green')
plt.title("Difference between normal and binomial")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show
```

## POISSON DISTRIBUTION

```python
x = random.poisson(lam = 2, size = 10)
print(x)
```

## VISUALIZATION OF POISSON DISTRIBUTION

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
poisson_data = random.poisson(lam = 2, size = 10)
sns.histplot(poisson_data, bins = 10, kde = False, color = 'skyblue')
plt. title ("A poison Distribution")
plt.xlabel ('Number of success')
plt.ylabel ('Frequency')
plt.show
```
